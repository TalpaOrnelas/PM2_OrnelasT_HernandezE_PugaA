### 1.4 Modelo que representa el problema.
> El modelo utilizado en este proyecto es el algoritmo de Diamond-Square, que es una técnica de generación de terrenos fractales. El algoritmo utiliza el concepto de interpolación para generar una matriz bidimensional representando un terreno. El terreno se genera iterativamente dividiendo la matriz en cuadrados y diamantes y actualizando los valores mediante la adición de un ruido aleatorio. La rugosidad y la altura se utilizan como parámetros para controlar el nivel de detalle y la variación en la elevación del terreno generado.

> Como se mencionó en este proyecto, se empleará el algoritmo de Diamond-Square en Python para crear un paisaje montañoso y se visualizará en forma de mapa, que se presenta a continuación:

> 1. Se importan los paquetes necesarios, como `numpy` y `matplotlib`, para manejar matrices y visualizar el paisaje, respectivamente.

> 2. Se definen los parámetros iniciales, como el tamaño del mapa (`size`), la rugosidad inicial (`roughness`), y el parámetro `h` que controla la rugosidad general del terreno.

> 3. Se crea una matriz inicial `landscape` de tamaño `(size, size)` llena de ceros, que posteriormente será utilizada para almacenar los valores del paisaje generado.

> 4. La función `normalize` se define para normalizar los valores de la matriz a un rango específico. Esto asegura que los valores se encuentren en el rango [0, 1], lo que facilita la visualización.

> 5. La función principal `diamond_square` implementa el algoritmo Diamond-Square. Este algoritmo divide repetidamente los cuadrados y los diamantes en la matriz, ajustando los valores de los puntos intermedios según un cálculo que involucra la rugosidad.

> 6. Una vez que se aplica el algoritmo `diamond_square` a la matriz `landscape`, los valores de la matriz se normalizan utilizando la función `normalize`.

> 7. Finalmente, se muestra el paisaje montañoso utilizando `matplotlib` y se utiliza el mapa de colores 'terrain' para visualizar el terreno con una apariencia similar a un paisaje natural.

> Ajusta los parámetros iniciales y experimenta con diferentes valores de `size`, `roughness` y `h` para crear variaciones en el paisaje montañoso generado. Esto te permitirá explorar diferentes tipos de paisajes montañosos fractales y su apariencia visual.

```python
import numpy as np
import matplotlib.pyplot as plt

# Tamaño del mapa
size = 129
# Parámetros iniciales
h = 0.5
roughness = 0.5

# Crear un mapa de tamaño (size, size) con valores iniciales aleatorios
paisaje = np.zeros((size, size))
```

### 1.5 Solución del problema.
> Gracias al algoritmo de Diamond-Square se proporcionará una solución factible para la generación de terrenos aleatorios. Ajustando los parámetros de rugosidad y altura, lograremos generar paisajes montañosos con diferentes niveles de detalle y variaciones. La solución obtenida es efectiva para simular terrenos naturales y ha demostrado ser una técnica robusta y versátil para la generación de paisajes en entornos virtuales y aplicaciones de simulación.

```python
# Función para normalizar los valores en el rango [0, 1]
def normalize(matrix):
    return (matrix - np.min(matrix)) / (np.max(matrix) - np.min(matrix))

# Función para aplicar el algoritmo Diamond-Square
def diamond_square(matrix, step_size, h, roughness):
    size = len(matrix) - 1
    half_step = step_size // 2
    if half_step < 1:
        return

    for y in range(half_step, size, step_size):
        for x in range(half_step, size, step_size):
            square_avg = np.mean([
                matrix[y - half_step, x - half_step], 
                matrix[y - half_step, x + half_step], 
                matrix[y + half_step, x - half_step], 
                matrix[y + half_step, x + half_step]
            ])
            matrix[y, x] = square_avg + (np.random.random() * 2 - 1) * roughness * step_size

    for y in range(0, size + 1, half_step):
        for x in range((y + half_step) % step_size, size + 1, step_size):
            diamond_avg = np.mean([
                matrix[(y - half_step) % size, x],
                matrix[(y + half_step) % size, x],
                matrix[y, (x - half_step) % size],
                matrix[y, (x + half_step) % size]
            ])
            matrix[y, x] = diamond_avg + (np.random.random() * 2 - 1) * roughness * step_size

    diamond_square(matrix, step_size // 2, h, roughness * 2 ** (-h))

    # Aplicar el algoritmo Diamond-Square
diamond_square(paisaje, size - 1, h, roughness)
```

### 1.6 Visualización de la solución del problema.
>Se ha generado una visualización gráfica del paisaje montañoso utilizando la biblioteca Matplotlib en Python. El mapa generado se representa en tonos de color que representan la elevación, lo que proporciona una representación visual clara de la topografía del terreno. Además, se ha incluido un gráfico de color para referencia de la altura. Los ejes de la gráfica se han etiquetado para una mejor comprensión de la representación visual del terreno.

```python
# Normalizar los valores del mapa a un rango [0, 1]
paisaje = normalize(paisaje)

# Mostrar el paisaje montañoso
plt.imshow(paisaje, cmap='terrain', origin='lower')
plt.colorbar()
plt.title('Paisaje Montañoso Fractal')
plt.show()
```

```python
# A Python translation of the Brownian Surface function from
# http://www.mathworks.com/matlabcentral/fileexchange/38945-fractional-brownian-field-or-surface-generator
#
# More info, check https://en.wikipedia.org/wiki/Brownian_surface
#
# Example output: http://i.imgur.com/MKHxA6N.png

import numpy as np

#%%%%%%%%%%%%%%%%%%%%%%%%%%%%%%%%%%%%%%%%%%%%%%%%%%%%%%%%%%%%%%%%%%%%%%%
# embedding of covariance function on a [0,R]^2 grid
def rho(x,y,R,alpha):

    if alpha <= 1.5:
        # alpha=2*H, where H is the Hurst parameter
        beta = 0
        c2 = alpha/2
        c0 = 1-alpha/2
    else:
        # parameters ensure piecewise function twice differentiable
        beta = alpha*(2-alpha)/(3*R*(R**2-1))
        c2 = (alpha-beta*(R-1)**2*(R+2))/2
        c0 = beta*(R-1)**3+1-c2

    # create continuous isotropic function
    r = np.sqrt((x[0]-y[0])**2+(x[1]-y[1])**2)
    if r<=1:
        out=c0-r**alpha+c2*r**2
    elif r<=R:
        out=beta*(R-r)**3/r
    else:
        out=0

    return out, c0, c2

# The main control is the Hurst parameter: H should be between 0 and
# 1, where 0 is very noisy, and 1 is smoother.
def brownian_surface(N=1000, H=0.95):
    R = 2  # [0,R]^2 grid, may have to extract only [0,R/2]^2

    # size of grid is m*n; covariance matrix is m^2*n^2
    M = N

    # create grid for field
    tx = np.linspace(0, R, M)
    ty = np.linspace(0, R, N)
    rows = np.zeros((M,N))


    for i in range(N):
        for j in range(M):
            # rows of blocks of cov matrix
            rows[j,i] = rho([tx[i],ty[j]],
                            [tx[0],ty[0]],
                            R, 2*H)[0]

    BlkCirc_row = np.vstack(
        [np.hstack([rows, rows[:,-1:1:-1]]),
         np.hstack([rows[-1:1:-1,:], rows[-1:1:-1, -1:1:-1]])])

    # compute eigen-values
    lam = np.real(np.fft.fft2(BlkCirc_row))/(4*(M-1)*(N-1))
    lam = np.sqrt(lam)

    # generate field with covariance given by block circular matrix
    Z = np.vectorize(complex)(np.random.randn(2*(M-1), 2*(M-1)),
                              np.random.randn(2*(M-1), 2*(M-1)))
    F = np.fft.fft2(lam*Z)
    F = F[:M, :N] # extract sub-block with desired covariance

    out,c0,c2 = rho([0,0],[0,0],R,2*H)

    field1 = np.real(F) # two independent fields
    field2 = np.imag(F)
    field1 = field1 - field1[0,0] # set field zero at origin
    field2 = field2 - field2[0,0] # set field zero at origin

    # make correction for embedding with a term c2*r^2
    field1 = field1 + np.kron(np.array([ty]).T * np.random.randn(),
                              np.array([tx]) * np.random.randn())*np.sqrt(2*c2)
    field2 = field2 + np.kron(np.array([ty]).T * np.random.randn(),
                              np.array([tx])   * np.random.randn())*np.sqrt(2*c2)
    X,Y = np.meshgrid(tx,ty)

    field1 = field1[:N//2, :M//2]
    field2 = field2[:N//2, :M//2]
    return field1

if __name__=="__main__":
    import matplotlib.pyplot as plt
    from mpl_toolkits.mplot3d import Axes3D

    fig = plt.figure(1)
    ax = fig.add_subplot(111, projection='3d')
    field = brownian_surface(N=1000, H=0.95)
    X, Y = np.meshgrid(np.linspace(-1,1,field.shape[0]),
                       np.linspace(-1,1,field.shape[1]))
    ax.plot_surface(X, Y, field)
    # plt.imshow(field1)
    plt.title('Fractional Gaussian Field')
    plt.show()
```

# 2 Solucionar por medio de la aproximación montecarlo algún problema físico que se modele con integrales múltiples (dobles, triples, etc)

```python
# Escribir la función acá
def montecarlo1(f, a, b, n):
    # Generar N puntos uniformemente distribudos en el intervalo [a,b)
    ui = np.random.uniform(a, b, n)
    
    # Evaluar la función en los N puntos aleatorios, obtener el promedio y multiplicarlo por el intervalo
    integral = ((b-a)/n)*np.sum(f(ui))
    # integral = (b-a)*np.mean(f(ui))   Esta línea es equivalente a la anterior
    return integral
```

1. Como la parábola solo tiene una raiz es necesario utilizar esta formula 
$f(x) = a(x-2)^2 = ax^2$

A continuación sustituimos las coordenas (x,y) con las del punto (3.15, 6.2)
$y = a(x-0)^2$
$6.2 = a(3.15-0)^2$  
$6.2=9.9225a$

Por lo tanto, $a = 6.2/9.9225$

La ecuación de la párabola sería: $f(x) = 2480/3969x^2$

2. Uno de los clientes quiere la misma copa con la superficie externa pintada como se ve en la imagen (sin pintar el pedestal ni la base); cálculen que superficie deberá ser pintada.

Para poder calcular el área de superficie en revolución utilizaremos esta formula: $A= 2\pi \int_{x_f}^{x_i} f(x)\sqrt{1+(\frac{dy}{dx})^2 \text{dy}}$

De esta manera es necesario sustituir y derivar la función en términos de (y) para poder realizar la integral.

```python
def f1(x):
    return -(2*np.pi*np.sqrt(3969*x/2480)+0.640322621)
# 2. Definimos los limites de integracion
a, b = 6.2, 0

# 3. Definimos cantidades de terminos
N = np.logspace(0, 5, 6).astype(int)
```

```python
# Crear el DataFrame de pandas (tabla) que contendrá la información
tabla2 = pd.DataFrame(index=N, columns = ['Aprox f1'])

# Agregar nombre al índice para que aprezca en la tabla
tabla2.index.name = 'Cantidad de términos'

# Rellenar las celdas de la aproximación de cada función
for n in N:
    aprox_f1 = montecarlo1(f1, a, b, n)
    
    tabla2.loc[n, 'Aprox f1'] = aprox_f1
    
tabla2
        
```

De esta manera, es necesario pintar una superficie que se acerca a los 86 $cm^2$.

3. Otro cliente quiere que se le agreguen líneas decorativas en color dorado al exterior, calculen la longitud de las líneas si inician 4.5 cm por debajo del borde de la copa.

De acuerdo a los datos otorgados sabemos que las franjas se extienden 4.5 cm por debajo del borde de la copa. De esta forma le restamos a 6.2, 4.5).

Por lo tanto determinamos que los límites son (6.2, 1.7) superior e inferior respectivamente.
De esta manera, unicamente sustituimos los valores en la fórmula en terminos de "y".
En este caso pasamos directo a esta forma, ya que tenemos simplificado el cuadrado de la derivada de la función.
- $\int_{6.5}^{1.7} \sqrt{1+\frac {3969}{9920y}}\text{d}x$.

```python
def f1(x):
    return -np.sqrt(1+(3969/9920*x))
# 2. Definimos los limites de integracion
a, b = 6.2, 1.7

# 3. Definimos cantidades de terminos
N = np.logspace(0, 5, 6).astype(int)
```

```python
# Crear el DataFrame de pandas (tabla) que contendrá la información
tabla2 = pd.DataFrame(index=N, columns = ['Aprox f1'])

# Agregar nombre al índice para que aprezca en la tabla
tabla2.index.name = 'Cantidad de términos'

# Rellenar las celdas de la aproximación de cada función
for n in N:
    aprox_f1 = montecarlo1(f1, a, b, n)
    
    tabla2.loc[n, 'Aprox f1'] = aprox_f1
        
tabla2
```

Por lo tanto, la longitud de las líneas mencionadas se aproxima a los 7.2 cm de longitud.

4. Aproximen la capacidad de la copa (consideren despreciable el espesor de la pared).

La formula para calcular el volumen de un solido en revolución es: 
$V=\int_{x_f}^{x_i} \pi[f(x)]^2 \text{{d}x}$

Utilizaremos la formula en base a "y" ya que facilita su menejo considerando 6.2 y o como los límetes de integración superiore inferior.

```python
def f1(x):
    return -(np.pi*3969*x/2480)
# 2. Definimos los limites de integracion
a, b = 6.2, 0

# 3. Definimos cantidades de terminos
N = np.logspace(0, 5, 6).astype(int)
```

```python
# Crear el DataFrame de pandas (tabla) que contendrá la información
tabla2 = pd.DataFrame(index=N, columns = ['Aprox f1'])

# Agregar nombre al índice para que aprezca en la tabla
tabla2.index.name = 'Cantidad de términos'

# Rellenar las celdas de la aproximación de cada función
for n in N:
    aprox_f1 = montecarlo1(f1, a, b, n)
    
    tabla2.loc[n, 'Aprox f1'] = aprox_f1
        
tabla2
```
