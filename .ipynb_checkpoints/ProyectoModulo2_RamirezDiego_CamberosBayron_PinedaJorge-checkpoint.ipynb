## 1.3 Modelo que representa el problema.



El modelo que representa el problema de optimización de procesos industriales utilizando la simulación de Monte Carlo es una herramienta precisa y flexible para analizar el desempeño de los procesos en condiciones de incertidumbre. Este modelo se basa en la generación de escenarios aleatorios que reflejan las variables de entrada del proceso, considerando su variabilidad y distribución de probabilidad.

A través de la simulación de Monte Carlo, se generan datos de salida que permiten evaluar el desempeño del proceso en cada escenario simulado. Estos datos incluyen medidas de desempeño como tiempos de producción, costos, utilización de recursos y calidad del producto. El análisis estadístico de estos datos proporciona información sobre patrones, tendencias y relaciones en el desempeño del proceso.

El modelo de simulación de Monte Carlo permite identificar las fuentes de incertidumbre más significativas que afectan el proceso, así como evaluar el impacto de diferentes estrategias y variables de decisión. A través del análisis de sensibilidad y la optimización, se pueden determinar las variables más influyentes y proponer estrategias de mejora para optimizar la eficiencia y la calidad del proceso.

### Problema

Somos una constructora que estamos en proyecto de construcción de un edificio de oficinas. El proyecto consta de las siguientes actividades principales:

>-Diseño arquitectónico: 6 semanas

>-Aprobación de permisos y licencias: 4 semanas

>-Preparación del sitio: 3 semanas

>-Excavación y cimentación: 5 semanas

>-Estructura de acero: 10 semanas

>-Albañilería y acabados: 8 semanas

>-Instalaciones eléctricas y de plomería: 6 semanas

>-Instalaciones de HVAC (calefacción, ventilación y aire acondicionado): 4 semanas

>-Instalaciones de sistemas de seguridad: 2 semanas

>-Pruebas y puesta en marcha: 3 semanas

>-Inspección final y entrega: 1 semana

Se sabe que existen incertidumbres en la duración de cada actividad debido a factores como el clima, la disponibilidad de recursos, cambios en los requisitos del cliente, entre otros. A continuación, se proporcionan las distribuciones de probabilidad para la duración de cada actividad:

>-Diseño arquitectónico: Distribución normal con media 6 semanas y desviación estándar 1 semana.

>-Aprobación de permisos y licencias: Distribución triangular con mínimo 3 semanas, moda 4 semanas y máximo 5 semanas. 

>-Preparación del sitio: Distribución triangular con mínimo 2 semanas, moda 3 semanas y máximo 4 semanas.

>-Excavación y cimentación: Distribución normal con media 5 semanas y desviación estándar 1 semana. 

>-Estructura de acero: Distribución uniforme entre 8 y 12 semanas. 

>-Albañilería y acabados: Distribución triangular con mínimo 7 semanas, moda 8 semanas y máximo 10 semanas. 

>-Instalaciones eléctricas y de plomería: Distribución normal con media 6 semanas y desviación estándar 0.5 semanas.

>-Instalaciones de HVAC: Distribución triangular con mínimo 3 semanas, moda 4 semanas y máximo 5 semanas. 

>-Instalaciones de sistemas de seguridad: Distribución uniforme entre 1 y 3 semanas.

>-Pruebas y puesta en marcha: Distribución triangular con mínimo 2 semanas, moda 3 semanas y máximo 4 semanas. 

>-Inspección final y entrega: Distribución normal con media 1 semana y desviación estándar 0.25 semanas.

Con estos datos que se nos proporciona utilizaremos simulación de Monte Carlo, realizando 1000 iteraciones para estimar la duración total del proyecto y el rango probable de tiempo requerido para completarlo.  

En cada iteración, debes muestrear aleatoriamente la duración de cada actividad según su distribución de probabilidad correspondiente. Al final de cada iteración, suma las duraciones de todas las actividades para obtener la duración total del proyecto.

## 1.5 Visualización de la solución dle problema.



```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
distribuciones = [
    ("Diseño arquitectónico", "normal", 6, 1),
    ("Aprobación de permisos y licencias", "triangular", 3, 4, 5),
    ("Preparación del sitio", "triangular", 2, 3, 4),
    ("Excavación y cimentación", "normal", 5, 1),
    ("Estructura de acero", "uniforme", 8, 12),
    ("Albañilería y acabados", "triangular", 7, 8, 10),
    ("Instalaciones eléctricas y de plomería", "normal", 6, 0.5),
    ("Instalaciones de HVAC", "triangular", 3, 4, 5),
    ("Instalaciones de sistemas de seguridad", "uniforme", 1, 3),
    ("Pruebas y puesta en marcha", "triangular", 2, 3, 4),
    ("Inspección final y entrega", "normal", 1, 0.25)]

num_iteraciones = 1000
duraciones_proyecto = []

for _ in range(num_iteraciones):
    duraciones_actividades = []
    for actividad, distribucion, *parametros in distribuciones:
        if distribucion == "normal":
            duracion = np.random.normal(*parametros)
        elif distribucion == "uniforme":
            duracion = np.random.uniform(*parametros)
        elif distribucion == "triangular":
            duracion = np.random.triangular(*parametros)
        else:
            raise ValueError("Distribución no válida.")
        duraciones_actividades.append(duracion)
    
    duracion_total = sum(duraciones_actividades)
    duraciones_proyecto.append(duracion_total)
```

```python
media_duracion = np.mean(duraciones_proyecto)
rango_probable = np.percentile(duraciones_proyecto, [5, 95])
duracion_minima = np.min(duraciones_proyecto)
duracion_maxima = np.max(duraciones_proyecto)
```

```python
print("Duración mínima del proyecto: {:.2f} semanas".format(duracion_minima))
print("Duración máxima del proyecto: {:.2f} semanas".format(duracion_maxima))
print("Duración promedio del proyecto: {:.2f} semanas".format(media_duracion))
print("Rango probable de tiempo requerido para completar el proyecto (percentil 5-95%): {:.2f} - {:.2f} semanas".format(rango_probable[0], rango_probable[1]))
```

```python
probabilidad_50_semanas = sum(duracion <= 50 for duracion in duraciones_proyecto) / num_iteraciones
print("Probabilidad de completar el proyecto en 50 semanas: {:.2%}".format(probabilidad_50_semanas))
```

Basándonos en los resultados obtenidos de la simulación Monte Carlo, podemos inferir que la duración del proyecto se encuentra en un rango de 45.28 a 60.21 semanas, con una duración promedio de 52.27 semanas. Esto indica que existe una variabilidad significativa en las duraciones posibles del proyecto.

Al analizar el rango probable de tiempo requerido para completar el proyecto, que se encuentra entre el percentil 5 y el percentil 95 (48.77 - 55.80 semanas), podemos observar que la gran mayoría de las duraciones del proyecto se encuentran dentro de este rango. Esto nos brinda una idea más precisa de los posibles escenarios y nos permite establecer límites probables para la finalización del proyecto.

Estos resultados son de gran importancia para la gestión del proyecto, ya que proporcionan información cuantitativa sobre la duración del proyecto y la variabilidad asociada. Esto permite tomar decisiones informadas y establecer expectativas realistas en términos de cronograma y entrega del proyecto.

En resumen, los resultados  indican que el proyecto puede tomar un tiempo considerable para completarse, con una variabilidad significativa en las duraciones posibles. Sin embargo, al considerar el rango probable de tiempo requerido para completar el proyecto, podemos establecer límites más realistas y tomar acciones para mitigar los posibles retrasos. Estos resultados son valiosos para la gestión y planificación efectiva del proyecto en el campo de la ingeniería civil.

```python
# Grafico de dispersión para visualizar las duraciones 
plt.figure(figsize=(10, 6))
plt.scatter(range(num_iteraciones), duraciones_proyecto, color='green', alpha=0.5)
plt.xlabel('Iteración')
plt.ylabel('Duración del proyecto (semanas)')
plt.title('Duraciones del proyecto en cada iteración')
plt.grid()
plt.show()
```

```python
# Grafico de densidad de probabilidad
import seaborn as sns

plt.figure(figsize=(10, 6))
sns.kdeplot(duraciones_proyecto, color='blue', fill=True)
plt.xlabel('Duración del proyecto (semanas)')
plt.ylabel('Densidad de probabilidad')
plt.title('Distribución de las duraciones del proyecto')
plt.grid()
plt.show()
```

```python
# Grafico de Evolucion con 100 iteraciones
num=100
duraciones_proyecto = []

for _ in range(num):
    duraciones_actividades = []
    for actividad, distribucion, *parametros in distribuciones:
        if distribucion == "normal":
            duracion = np.random.normal(*parametros)
        elif distribucion == "uniforme":
            duracion = np.random.uniform(*parametros)
        elif distribucion == "triangular":
            duracion = np.random.triangular(*parametros)
        else:
            raise ValueError("Distribución no válida.")
        duraciones_actividades.append(duracion)
    
    duracion_total = sum(duraciones_actividades)
    duraciones_proyecto.append(duracion_total)

plt.figure(figsize=(10, 6))
plt.plot(range(num), duraciones_proyecto, color='blue')
plt.xlabel('Iteraciones')
plt.ylabel('Duración del proyecto (semanas)')
plt.title('Evolución de las duraciones del proyecto')
plt.grid()
plt.show()
```

Al analizar los resultados gráficos obtenidos a través de la simulación Monte Carlo, se puede obtener una visión más completa y visual de las duraciones del proyecto en diferentes escenarios.

El histograma generado muestra la distribución de las duraciones del proyecto. Se observa que las duraciones se agrupan alrededor de un valor central, lo que indica que es más probable que el proyecto se complete en torno a esa duración promedio. Además, la forma del histograma proporciona información sobre la dispersión de las duraciones y la probabilidad de duraciones extremas.

El gráfico de líneas muestra la evolución de las duraciones del proyecto a lo largo de las iteraciones. Esto permite observar cómo varían las duraciones en cada simulación y si hay alguna tendencia o patrón discernible. Si la línea muestra una trayectoria estable o converge hacia un valor específico, indica una mayor consistencia en las duraciones del proyecto a medida que se realizan más iteraciones.

Estos resultados gráficos complementan los resultados numéricos al proporcionar una representación visual de las duraciones del proyecto. Esto facilita la comprensión de la distribución de duraciones, la variabilidad y las posibles tendencias en los datos simulados.

```python
duracion_minima = 45.28
duracion_maxima = 60.21
duracion_promedio = 52.27
duracion_total = 60 

porcentaje_incremento = [(i - duracion_minima) / (duracion_maxima - duracion_minima) * 100
                          for i in np.linspace(duracion_minima, duracion_promedio, duracion_total)]

#Gráfica de porcentaje de incremento según el avance de las semanas
plt.figure(figsize=(10, 6))
plt.plot(range(1, duracion_total + 1), porcentaje_incremento, color='orange')
plt.xlabel('Semana')
plt.ylabel('Porcentaje de terminar proyecto')
plt.title('Porcentaje de Incremento según el Avance de las Semanas')
plt.grid()
plt.show()
```

Podemos observar a través del gráfico que a medida que avanzan las semanas, existe una mayor probabilidad de que el proyecto se concluya al 100%. Esto se debe a que el porcentaje de terminar el proyecto aumenta gradualmente a medida que transcurre el tiempo.
Inicialmente, en las primeras semanas, el porcentaje de terminar el proyecto es bajo, ya que el tiempo transcurrido es insuficiente para completar la totalidad del proyecto. Sin embargo, a medida que avanzan las semanas, el porcentaje va incrementando, indicando que se está progresando en la finalización del proyecto.
Cuando se acerca a las últimas semanas del proyecto, el porcentaje de terminar el proyecto se acerca al 100%. Esto implica que, con base en las duraciones mínima, máxima y promedio establecidas, existe una alta probabilidad de que el proyecto esté completamente concluido al final de la duración prevista.
